## (a)

Closed form expression of $\mathbb{E} Z \mathbf{1}_{|Z| \leq \gamma}$ :
$$
\begin{aligned}
\mathbb{E} Z \mathbf{1}_{|Z| \leq \gamma} & =\int_{-\infty}^{\infty} \frac{z}{\tau \sqrt{2 \pi}} e^{-\frac{1}{2}\left(\frac{z-\omega}{\tau}\right)^2} \mathbf{1}_{|Z| \leq \gamma} d z \\
& =\frac{1}{\sqrt{2 \pi}} \int_{-\gamma}^\gamma \frac{z}{\tau} e^{-\frac{1}{2}\left(\frac{z-\omega}{\tau}\right)^2} d z
\end{aligned}
$$

## (b)
(b) Closed form expression of $\mathbb{E}\left(X-\hat{\mu}_n^H\right)^2$ :
$$
\begin{aligned}
\mathbb{E}\left(X-\hat{\mu}_n^H\right)^2 & =\mathbb{E}\left((X-\mu)+\left(\mu-\hat{\mu}_n^H\right)\right)^2 \\
& =\mathbb{E}(X-\mu)^2+\mathbb{E}\left(\mu-\hat{\mu}_n^H\right)^2 \\
& =\sigma^2+\mathbb{E}\left(\mu-\hat{\mu}_n^H\right) \\
& =\sigma^2+\mathbb{E}\left(\mu^2-2 \mu \hat{\mu}_n^H+\hat{\mu}_n^{H 2}\right) \\
& =\sigma^2+\mu^2-2 \mu\left(\mathbb{E} \bar{X}_n-\mathbb{E} \bar{X}_n \mathbf{1}_{\left|\bar{X}_n\right| \leq \alpha}\right)+\mathbb{E} \hat{\mu}_n^{H 2} \\
& =\sigma^2+\mu^2-2 \mu^2+2 \mu \mathbb{E} \bar{X}_n \mathbf{1}_{\left|\bar{X}_n\right| \leq \alpha}+\mathbb{E} \hat{\mu}_n^{H 2} \\
& =\sigma^2-\mu^2+2 \mu \mathbb{E} \bar{X}_n \mathbf{1}_{\left|\bar{X}_n\right| \leq \alpha}+\mathbb{E} \hat{\mu}_n^H{ }^2 \\
& =\sigma^2-\mu^2+2 \mu \mathbb{E} \bar{X}_n \mathbf{1}_{\left|\bar{X}_n\right| \leq \alpha}+\mathbb{E}\left(\bar{X}_n^2-2 \bar{X}_n\left(\bar{X}_n \mathbf{1}_{\left|\bar{X}_n\right| \leq \alpha}\right)+\left(\bar{X}_n \mathbf{1}_{\left|\bar{X}_n\right| \leq \alpha}\right)^2\right) \\
& =\sigma^2-\mu^2+2 \mu \mathbb{E} \bar{X}_n \mathbf{1}_{\left|\bar{X}_n\right| \leq \alpha}+\mathbb{E}\left(\bar{X}_n^2\right)+\mathbb{E}\left(-2 \bar{X}_n\left(\bar{X}_n \mathbf{1}_{\left|\bar{X}_n\right| \leq \alpha}\right)+\left(\bar{X}_n \mathbf{1}_{\left|\bar{X}_n\right| \leq \alpha}\right)^2\right) \\
& =\sigma^2-\mu^2+2 \mu \mathbb{E} \bar{X}_n \mathbf{1}_{\left|\bar{X}_n\right| \leq \alpha}+\frac{\sigma^2}{n}+\mu^2+\mathbb{E}\left(-2 \bar{X}_n\left(\bar{X}_n \mathbf{1}_{\left|\bar{X}_n\right| \leq \alpha}\right)+\left(\bar{X}_n \mathbf{1}_{\left|\bar{X}_n\right| \leq \alpha}\right)^2\right) \\
& =\sigma^2+2 \mu \mathbb{E} \bar{X}_n \mathbf{1}_{\left|\bar{X}_n\right| \leq \alpha}+\frac{\sigma^2}{n}-\mathbb{E}\left(\bar{X}_n^2 \mathbf{1}_{\left|\bar{X}_n\right| \leq \alpha}\right)
\end{aligned}
$$
We can plug the closed form from (a) in for $\mathbb{E} \bar{X}_n \mathbf{1}_{\left|\bar{X}_n\right| \leq \alpha}$ and evaluate the second moment of $\bar{X}_n \mathbf{1}_{\left|\bar{X}_n\right| \leq \alpha}$ to obtain the final expression. To solve for $\alpha^{\text {opt }}$, take the derivative with respect to $\alpha$ and set the resulting expression $=0$. One can also perform a computational search across a range of $\alpha$ values to identify an approximate optimal value.

## (c)
A possible estimator of $\alpha^{\text {opt }}$ would be
$$
\hat{\alpha}_n=\underset{\alpha}{\arg \min }\left(\text { closed form of } \mathbb{E}\left(X-\hat{\mu}_n^H\right)^2\right)
$$
where $\mu$ and $\sigma^2$ are replaced with the sample mean $\left(\bar{X}=\frac{1}{n} \sum_{i=1}^n X_i\right)$ and variance, given by
$$
\frac{1}{n-1} \sum_{i=1}^n\left(X_i-\bar{X}\right)^2

## (a)

Since $\epsilon$ and $\mathbf{X}$ are independent we see that:
$$
\begin{aligned}
\operatorname{Var}(Y) & =\operatorname{Var}\left(\mathbf{X}^T \beta^*+\epsilon\right) \\
& =\operatorname{Var}\left(\mathbf{X}^T \beta^*\right)+\operatorname{Var}(\epsilon) \\
& =\beta^{* T} \Sigma(\rho) \beta^*+\sigma^2
\end{aligned}
$$

Also, $P\left(Y-\mathbf{X}^T \beta^*\right)^2=P \epsilon^2$ as the expected squared residuals. Since $P \epsilon=0$ then $P \epsilon^2=\sigma^2$ and $P\left(Y-\mathbf{X}^T \beta^*\right)^2=\sigma^2$.
Thus,
$$
\begin{gathered}
R^2=1-\frac{P\left(Y-\mathbf{X}^T \beta^*\right)^2}{\operatorname{Var}(Y)} \\
=1-\frac{\sigma^2}{\beta^{* T} \Sigma(\rho) \beta^*+\sigma^2} \\
\sigma^2=\left(\sigma^2+\beta^{* T} \Sigma(\rho) \beta^*\right)\left(1-R^2\right) \\
\sigma^2=\left(\sigma^2\right)\left(1-R^2\right)+\left(\beta^{* T} \Sigma(\rho) \beta^*\right)\left(1-R^2\right) \\
\sigma^2 R^2=\left(\beta^{* T} \Sigma(\rho) \beta^*\right)\left(1-R^2\right) \\
\sigma^2=\frac{\left(\beta^{* T} \Sigma(\rho) \beta^*\right)\left(1-R^2\right)}{R^2}
\end{gathered}
$$

Let $R^2=r$ and your final equation is:
$$
\sigma^2=\frac{1-r}{r}\left(\beta^{* T} \Sigma(\rho) \beta^*\right)
$$

## (b)
When $R^2=0.8, \sigma^2$ is calculated as:
$$
\begin{aligned}
0.8 & =1-\frac{\sigma^2}{\beta^{* T} \Sigma(\rho) \beta^*+\sigma^2} \\
0.2 & =\frac{\sigma^2}{\beta^{* T} \Sigma(\rho) \beta^*+\sigma^2} \\
\sigma^2 & =0.2\left(\beta^{* T} \Sigma(\rho) \beta^*+\sigma^2\right) \\
0.8 \sigma^2 & =0.2\left(\beta^{* T} \Sigma(\rho) \beta^*\right) \\
\sigma^2 & =0.25\left(\beta^{* T} \Sigma(\rho) \beta^*\right)
\end{aligned}
$$

```python
def calc_aic (y_hat, y, D, lamb):
    df_1 = np.sum(D**2/(D**2 + lamb))
    rss = np. sum((y.ravel() - y_hat.ravel())**2)
    aic = np.log(rss) + df_1*2/len(y)
    return aic

def calc_bic(y_hat, y, D, lamb):
    df_1 = np.sum(D**2/(D**2 + lamb))
    rss = np.sum((y - y_hat)**2)
    bic = np.log(rss) + df_1*np.log(len(y))/len(y)
    return bic

def lambda_ic (lambs, X, Y, X_star, Y_star, ic='aic'):
    U, D, V = np.linalg.svd(X, full_matrices=False)
    y_pred = lambda l: (U @ np.diag(D**2/(D**2 + l)) @ U.T) @ Y
    if ic=='bic':
        bic_vals = np.array([calc_bic(y_hat = y_pred(l), y=Y, D=D, lamb=l) for l in lambs])
        min_i = np.argmin(bic_vals)
    else:
        aic_vals = np.array([calc_aic(y_hat = y_pred(l), y=Y, D=D, lamb=l) for l in lambs])
        min_i = np.argmin(aic_vals)
    mse = mean_squared_error(Y, y_pred(lambs[min_i]))
    ridge = Ridge(alpha=lambs[i], fit_intercept=False)
    ridge.fit(X,Y)
    omse = mean_squared_error(Y_star, ridge.predict(X_star))
    return mse, lambs[min_i], omse
```

```python
EMSES = np.zeros((24, 9))

datasets = 1000
n = 100
count = 0
test_size = 1000

# Sparse Signal
lasso_aic_mse_spr = np.zeros(9)
lasso_bic_mse_spr = np.zeros(9)
lasso_loo_mse_spr = np.zeros(9)
adp_lasso_aic_mse_spr = np.zeros(9)
adp_lasso_bic_mse_spr = np.zeros(9)
adp_lasso_loo_mse_spr = np.zeros(9)

ridge_aic_mse_spr = np.zeros(9)
ridge_bic_mse_spr = np.zeros(9)
ridge_loo_mse_spr = np.zeros(9)
adp_ridge_aic_mse_spr = np.zeros(9)
adp_ridge_bic_mse_spr = np.zeros(9)
adp_ridge_loo_mse_spr = np.zeros(9)

#Dense Signal
lasso_aic_mse_dns = np.zeros(9)
lasso_bic_mse_dns = np.zeros(9)
lasso_loo_mse_dns = np.zeros(9)
adp_lasso_aic_mse_dns = np.zeros(9)
adp_lasso_bic_mse_dns = np.zeros(9)
adp_lasso_loo_mse_dns = np.zeros(9)

ridge_aic_mse_dns = np.zeros(9)
ridge_bic_mse_dns = np.zeros(9)
ridge_loo_mse_dns = np.zeros(9)
adp_ridge_aic_mse_dns = np.zeros(9)
adp_ridge_bic_mse_dns = np.zeros(9)
adp_ridge_loo_mse_dns = np.zeros(9)

for i, (p, rho) in enumerate(itertools.product([10, 25, 50], [0, 0.25, 0.5])):
    print('p = ', p, 'rho = ', rho)
    
    # Generate data
    beta_spr = np.fromfunction(lambda j, _: (2/np.sqrt(n)) * (j + 1 <= np.sqrt(p)), (p, 1))
    beta_dns = np.fromfunction(lambda j, _: (5/((j+1) * np.sqrt(n))), (p, 1))
    Sigma_p = np.fromfunction(lambda i, j: rho**(np.abs(i -j)), (p, p))
    sigma2_spr = 0.25 * (beta_spr.T @ Sigma_p @ beta_spr)
    sigma2_dns = 0.25 * (beta_dns.T @ Sigma_p @ beta_dns)
    
    # Generate test sets
    test_epsilon_spr = np.random.normal(loc = 0, scale = np.sqrt(sigma2_spr))
    test_epsilon_dns = np.random.normal(loc = 0, scale = np.sqrt(sigma2_dns))
    
    test_X = np.random.multivariate_normal(mean = np.zeros(p), cov = Sigma_p, size = test_size)
    
    test_Y_spr = test_X @ beta_spr + test_epsilon_spr
    test_Y_dns = test_X @ beta_dns + test_epsilon_dns
    
    # LOOCV
    loo = LeaveOneOut()
    prog = tqdm(total = datasets, position = 0, leave = 2)
    for _ in range(datasets):
        count += 1
        
        epsilon_spr = np.random.normal(loc = 0, scale = np.sqrt(sigma2_spr))
        epsilon_dns = np.random.normal(loc = 0, scale = np.sqrt(sigma2_dns))
        
        X = np.random.multivariate_normal(mean = np.zeros(p), cov = Sigma_p, size = n)
        
        Y_spr = X @ beta_spr + epsilon_spr
        Y_dns = X @ beta_dns + epsilon_dns
        
        lm_spr = LinearRegression().fit(X, Y_spr)
        lm_dns = LinearRegression().fit(X, Y_dns)
        
        # Adaptive Lasso
        adX_sprL = X @ np.diagflat(np.abs(lm_spr.coef_))
        adX_dnsL = X @ np.diagflat(np.abs(lm_dns.coef_))
        
        # Adaptive Ridge
        adX_sprR = X @ np.diagflat((lm_spr.coef_**2)**(1/2))
        adX_dnsR = X @ np.diagflat((lm_dns.coef_**2)**(1/2))
        
        # Adaptive Lasso
        test_adX_sprL = test_X @ np.diagflat(np.abs(lm_spr.coef_))
        test_adX_dnsL = test_X @ np.diagflat(np.abs(lm_dns.coef_))
        
        # Adaptive Ridge
        test_adX_sprR = test_X @ np.diagflat((lm_spr.coef_**2)**(1/2))
        test_adX_dnsR = test_X @ np.diagflat((lm_dns.coef_**2)**(1/2))
        
        # Calculate lasso
        # AIC
        # Sparse
        lasso_aic_spr = LassoLarsIC(criterion='aic')
        lasso_aic_spr.fit(X, Y_spr)
        lasso_aic_mse_spr[i] += mean_squared_error(test_Y_spr, lasso_aic_spr.predict(test_X))
        
        # Dense
        lasso_aic_dns = LassoLarsIC(criterion='aic')
        lasso_aic_dns.fit(X, Y_dns)
        lasso_aic_mse_dns[i] += mean_squared_error(test_Y_dns, lasso_aic_dns.predict(test_X))
        
        # BIC
        # Sparse
        lasso_bic_spr = LassoLarsIC(criterion='bic')
        lasso_bic_spr.fit(X, Y_spr)
        lasso_bic_mse_spr[i] += mean_squared_error(test_Y_spr, lasso_bic_spr.predict(test_X))
        
        # Dense
        lasso_bic_dns = LassoLarsIC(criterion='bic')
        lasso_bic_dns.fit(X, Y_dns)
        lasso_bic_mse_dns[i] += mean_squared_error(test_Y_dns, lasso_bic_dns.predict(test_X))
        
        # LOOCV
        # Sparse
        lasso_loo_spr = LassoLarsCV(cv=loo)
        lasso_loo_spr.fit(X, Y_spr)
        lasso_loo_mse_spr[i] += mean_squared_error(test_Y_spr, lasso_loo_spr.predict(test_X))
        
        # Dense
        lasso_loo_dns = LassoLarsCV(cv=loo)
        lasso_loo_dns.fit(X, Y_dns)
        lasso_loo_mse_spr[i] += mean_squared_error(test_Y_spr, lasso_loo_spr.predict(test_X))
        
        
        # Calculate adaptive lasso
        # AIC
        # Sparse
        adp_lasso_aic_spr = LassoLarsIC(criterion='aic')
        adp_lasso_aic_spr.fit(adX_sprL, Y_spr)
        adp_lasso_aic_mse_spr[i] += mean_squared_error(test_Y_spr, adp_lasso_aic_spr.predict(test_X))
        
        # Dense
        adp_lasso_aic_dns = LassoLarsIC(criterion='aic')
        adp_lasso_aic_dns.fit(adX_dnsL, Y_dns)
        adp_lasso_aic_mse_dns[i] += mean_squared_error(test_Y_dns, adp_lasso_aic_dns.predict(test_X))
        
        # BIC
        # Sparse
        adp_lasso_bic_spr = LassoLarsIC(criterion='bic')
        adp_lasso_bic_spr.fit(adX_sprL, Y_spr)
        adp_lasso_bic_mse_spr[i] += mean_squared_error(test_Y_spr, adp_lasso_bic_spr.predict(test_X))
        
        # Dense
        adp_lasso_bic_dns = LassoLarsIC(criterion='bic')
        adp_lasso_bic_dns.fit(adX_dnsL, Y_dns)
        adp_lasso_bic_mse_dns[i] += mean_squared_error(test_Y_dns, adp_lasso_bic_dns.predict(test_X))
        
        # LOOCV
        # Sparse
        adp_lasso_loo_spr = LassoLarsCV(cv=loo)
        adp_lasso_loo_spr.fit(adX_sprL, Y_spr)
        adp_lasso_loo_mse_spr[i] += mean_squared_error(test_Y_spr, adp_lasso_loo_spr.predict(test_X))
        
        # Dense
        adp_lasso_loo_dns = LassoLarsCV(cv=loo)
        adp_lasso_loo_dns.fit(adX_dnsL, Y_dns)
        adp_lasso_loo_mse_spr[i] += mean_squared_error(test_Y_spr, adp_lasso_loo_spr.predict(test_X))
        
        # Calculat ridge
        # AIC
        lambda_rng = np.linspace(0.001, 100, num = 200)
        
        # Sparse
        _,__,mse = lambda_ic(lambda_rng, X, Y_spr, test_X, test_Y_spr, ic = 'aic')
        ridge_aic_mse_spr[i] += mse
        
        # Dense
        _,__,mse = lambda_ic(lambda_rng, X, Y_dns, test_X, test_Y_dns, ic = 'aic')
        ridge_aic_mse_dns[i] += mse
        
        # BIC
        
        # Sparse
        _,__,mse = lambda_ic(lambda_rng, X, Y_spr, test_X, test_Y_spr, ic = 'bic')
        ridge_bic_mse_spr[i] += mse
        
        # Dense
        _,__,mse = lambda_ic(lambda_rng, X, Y_dns, test_X, test_Y_dns, ic = 'bic')
        ridge_bic_mse_dns[i] += mse
        

        # LOOCV
        
        # Sparse
        ridge_loo_spr = RidgeCV(cv = loo)
        ridge_loo_spr.fit(X, Y_spr)
        ridge_loo_mse_spr[i] += mean_squared_error(test_Y_spr, ridge_loo_spr.predict(test_X))
        
        # Dense
        ridge_loo_dns = RidgeCV(cv = loo)
        ridge_loo_dns.fit(X, Y_dns)
        ridge_loo_mse_dns[i] += mean_squared_error(test_Y_dns, ridge_loo_dns.predict(test_X))
        
        
        # Calculat adaptive ridge
        # AIC

        # Sparse
        _,__,mse = lambda_ic(lambda_rng, adX_sprR, Y_spr, test_adX_sprR, test_Y_spr, ic = 'aic')
        adp_ridge_aic_mse_spr[i] += mse
        
        # Dense
        _,__,mse = lambda_ic(lambda_rng, adX_dnsR, Y_dns, test_adX_dnsR, test_Y_dns, ic = 'aic')
        adp_ridge_aic_mse_dns[i] += mse
        
        # BIC
        
        # Sparse
        _,__,mse = lambda_ic(lambda_rng, adX_sprR, Y_spr, test_adX_sprR, test_Y_spr, ic = 'bic')
        adp_ridge_bic_mse_spr[i] += mse
        
        # Dense
        _,__,mse = lambda_ic(lambda_rng, adX_dnsR, Y_dns, test_adX_dnsR, test_Y_dns, ic = 'bic')
        adp_ridge_bic_mse_dns[i] += mse
        

        # LOOCV
        
        # Sparse
        adp_ridge_loo_spr = RidgeCV(cv = loo)
        adp_ridge_loo_spr.fit(adX_sprR, Y_spr)
        adp_ridge_loo_mse_spr[i] += mean_squared_error(test_Y_spr, adp_ridge_loo_spr.predict(test_X))
        
        # Dense
        adp_ridge_loo_dns = RidgeCV(cv = loo)
        adp_ridge_loo_dns.fit(adX_dnsR, Y_dns)
        adp_ridge_loo_mse_dns[i] += mean_squared_error(test_Y_dns, adp_ridge_loo_dns.predict(test_X))
        
        prog.update()
```
