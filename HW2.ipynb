## (a)

Closed form expression of $\mathbb{E} Z \mathbf{1}_{|Z| \leq \gamma}$ :
$$
\begin{aligned}
\mathbb{E} Z \mathbf{1}_{|Z| \leq \gamma} & =\int_{-\infty}^{\infty} \frac{z}{\tau \sqrt{2 \pi}} e^{-\frac{1}{2}\left(\frac{z-\omega}{\tau}\right)^2} \mathbf{1}_{|Z| \leq \gamma} d z \\
& =\frac{1}{\sqrt{2 \pi}} \int_{-\gamma}^\gamma \frac{z}{\tau} e^{-\frac{1}{2}\left(\frac{z-\omega}{\tau}\right)^2} d z
\end{aligned}
$$

## (b)
(b) Closed form expression of $\mathbb{E}\left(X-\hat{\mu}_n^H\right)^2$ :
$$
\begin{aligned}
\mathbb{E}\left(X-\hat{\mu}_n^H\right)^2 & =\mathbb{E}\left((X-\mu)+\left(\mu-\hat{\mu}_n^H\right)\right)^2 \\
& =\mathbb{E}(X-\mu)^2+\mathbb{E}\left(\mu-\hat{\mu}_n^H\right)^2 \\
& =\sigma^2+\mathbb{E}\left(\mu-\hat{\mu}_n^H\right) \\
& =\sigma^2+\mathbb{E}\left(\mu^2-2 \mu \hat{\mu}_n^H+\hat{\mu}_n^{H 2}\right) \\
& =\sigma^2+\mu^2-2 \mu\left(\mathbb{E} \bar{X}_n-\mathbb{E} \bar{X}_n \mathbf{1}_{\left|\bar{X}_n\right| \leq \alpha}\right)+\mathbb{E} \hat{\mu}_n^{H 2} \\
& =\sigma^2+\mu^2-2 \mu^2+2 \mu \mathbb{E} \bar{X}_n \mathbf{1}_{\left|\bar{X}_n\right| \leq \alpha}+\mathbb{E} \hat{\mu}_n^{H 2} \\
& =\sigma^2-\mu^2+2 \mu \mathbb{E} \bar{X}_n \mathbf{1}_{\left|\bar{X}_n\right| \leq \alpha}+\mathbb{E} \hat{\mu}_n^H{ }^2 \\
& =\sigma^2-\mu^2+2 \mu \mathbb{E} \bar{X}_n \mathbf{1}_{\left|\bar{X}_n\right| \leq \alpha}+\mathbb{E}\left(\bar{X}_n^2-2 \bar{X}_n\left(\bar{X}_n \mathbf{1}_{\left|\bar{X}_n\right| \leq \alpha}\right)+\left(\bar{X}_n \mathbf{1}_{\left|\bar{X}_n\right| \leq \alpha}\right)^2\right) \\
& =\sigma^2-\mu^2+2 \mu \mathbb{E} \bar{X}_n \mathbf{1}_{\left|\bar{X}_n\right| \leq \alpha}+\mathbb{E}\left(\bar{X}_n^2\right)+\mathbb{E}\left(-2 \bar{X}_n\left(\bar{X}_n \mathbf{1}_{\left|\bar{X}_n\right| \leq \alpha}\right)+\left(\bar{X}_n \mathbf{1}_{\left|\bar{X}_n\right| \leq \alpha}\right)^2\right) \\
& =\sigma^2-\mu^2+2 \mu \mathbb{E} \bar{X}_n \mathbf{1}_{\left|\bar{X}_n\right| \leq \alpha}+\frac{\sigma^2}{n}+\mu^2+\mathbb{E}\left(-2 \bar{X}_n\left(\bar{X}_n \mathbf{1}_{\left|\bar{X}_n\right| \leq \alpha}\right)+\left(\bar{X}_n \mathbf{1}_{\left|\bar{X}_n\right| \leq \alpha}\right)^2\right) \\
& =\sigma^2+2 \mu \mathbb{E} \bar{X}_n \mathbf{1}_{\left|\bar{X}_n\right| \leq \alpha}+\frac{\sigma^2}{n}-\mathbb{E}\left(\bar{X}_n^2 \mathbf{1}_{\left|\bar{X}_n\right| \leq \alpha}\right)
\end{aligned}
$$
We can plug the closed form from (a) in for $\mathbb{E} \bar{X}_n \mathbf{1}_{\left|\bar{X}_n\right| \leq \alpha}$ and evaluate the second moment of $\bar{X}_n \mathbf{1}_{\left|\bar{X}_n\right| \leq \alpha}$ to obtain the final expression. To solve for $\alpha^{\text {opt }}$, take the derivative with respect to $\alpha$ and set the resulting expression $=0$. One can also perform a computational search across a range of $\alpha$ values to identify an approximate optimal value.

## (c)
A possible estimator of $\alpha^{\text {opt }}$ would be
$$
\hat{\alpha}_n=\underset{\alpha}{\arg \min }\left(\text { closed form of } \mathbb{E}\left(X-\hat{\mu}_n^H\right)^2\right)
$$
where $\mu$ and $\sigma^2$ are replaced with the sample mean $\left(\bar{X}=\frac{1}{n} \sum_{i=1}^n X_i\right)$ and variance, given by
$$
\frac{1}{n-1} \sum_{i=1}^n\left(X_i-\bar{X}\right)^2
$$

## (a)

Since $\epsilon$ and $\mathbf{X}$ are independent we see that:
$$
\begin{aligned}
\operatorname{Var}(Y) & =\operatorname{Var}\left(\mathbf{X}^T \beta^*+\epsilon\right) \\
& =\operatorname{Var}\left(\mathbf{X}^T \beta^*\right)+\operatorname{Var}(\epsilon) \\
& =\beta^{* T} \Sigma(\rho) \beta^*+\sigma^2
\end{aligned}
$$

Also, $P\left(Y-\mathbf{X}^T \beta^*\right)^2=P \epsilon^2$ as the expected squared residuals. Since $P \epsilon=0$ then $P \epsilon^2=\sigma^2$ and $P\left(Y-\mathbf{X}^T \beta^*\right)^2=\sigma^2$.
Thus,
$$
\begin{gathered}
R^2=1-\frac{P\left(Y-\mathbf{X}^T \beta^*\right)^2}{\operatorname{Var}(Y)} \\
=1-\frac{\sigma^2}{\beta^{* T} \Sigma(\rho) \beta^*+\sigma^2} \\
\sigma^2=\left(\sigma^2+\beta^{* T} \Sigma(\rho) \beta^*\right)\left(1-R^2\right) \\
\sigma^2=\left(\sigma^2\right)\left(1-R^2\right)+\left(\beta^{* T} \Sigma(\rho) \beta^*\right)\left(1-R^2\right) \\
\sigma^2 R^2=\left(\beta^{* T} \Sigma(\rho) \beta^*\right)\left(1-R^2\right) \\
\sigma^2=\frac{\left(\beta^{* T} \Sigma(\rho) \beta^*\right)\left(1-R^2\right)}{R^2}
\end{gathered}
$$

Let $R^2=r$ and your final equation is:
$$
\sigma^2=\frac{1-r}{r}\left(\beta^{* T} \Sigma(\rho) \beta^*\right)
$$

## (b)
When $R^2=0.8, \sigma^2$ is calculated as:
$$
\begin{aligned}
0.8 & =1-\frac{\sigma^2}{\beta^{* T} \Sigma(\rho) \beta^*+\sigma^2} \\
0.2 & =\frac{\sigma^2}{\beta^{* T} \Sigma(\rho) \beta^*+\sigma^2} \\
\sigma^2 & =0.2\left(\beta^{* T} \Sigma(\rho) \beta^*+\sigma^2\right) \\
0.8 \sigma^2 & =0.2\left(\beta^{* T} \Sigma(\rho) \beta^*\right) \\
\sigma^2 & =0.25\left(\beta^{* T} \Sigma(\rho) \beta^*\right)
\end{aligned}
$$
